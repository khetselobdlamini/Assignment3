### Age Range of Patients

This code creates a histogram of the "Age" column in the dataset "diabetes_df" by using a class called "HistPlot" and a function called "create_plot". The "create_plot" method requires three inputs: the plot's title ("Age range of patients"), the x-axis's ("Age") label, and the y-axis' ("Count"). The code first creates an instance of the "HistPlot" class for the "Age" column in "diabetes_df" (agerange = HistPlot(diabetes_df['Age'])), and then runs the "create_plot" method on that instance with the supplied title, x-label, and y-label. The resulting plot will probably display a histogram of the age distribution of the dataset's patients, with the y-axis reflecting the number of patients in each age range and the x-axis representing age ranges. The way the "HistPlot" class is used will determine how the plot exactly looks.

```python
#Age Range of Patients
class ScatterPlot:
    def __init__(self, x_data, y_data, color):
        self.x_data = x_data
        self.y_data = y_data
        self.color = color

    def create_plot(self, title, x_label, y_label):
        plt.scatter(x=self.x_data, y=self.y_data, color=self.color)
        plt.title(title)
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.grid(axis='both', alpha=0.75)
        plt.show()
        
class HistPlot:
    def __init__(self, x_data):
        self.x_data = x_data

    def create_plot(self, title, x_label, y_label):
        plt.figure(figsize = (8,4))
        plt.hist(x=self.x_data, linewidth = 2, edgecolor = 'black')
        plt.title(title)
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.grid(axis='both', alpha=0.75)
        plt.show()
        
agerange = HistPlot(diabetes_df['Age'])
agerange.create_plot('Age range of patients', 'Age', 'Count')
```

### Histograms for each Column

The distribution of features in the dataset is visualized using this code, which generates a grid of 2x4 subplots. It begins by removing the feature names from the dataset and adding them to a list called "features". The histogram of values is then calculated for each feature using Numpy, and it is plotted using Matplotlib. Additionally, it employs seaborn to overlay a kernel density graphic. Finally, it uses matplotlib to display the plot and sets the titles of each subplot to the respective feature names. Overall, this code gives a brief description of how each feature in the dataset is distributed. 

```python
features = diabetes_df.iloc[:, :-1].columns.to_list()

fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(15, 10),dpi=200)
for i, feature in enumerate(features):
    row = i // 4
    col = i % 4
    counts, bin_edges = np.histogram(diabetes_df[feature], density=True)
    axes[row, col].hist(diabetes_df[feature], bins=bin_edges, alpha=0.5, color="#A52A2A", density=True)
    sns.kdeplot(diabetes_df[feature], ax=axes[row, col], color='black', linewidth=1)
    axes[row, col].set_title(f'Histogram of {feature}')
plt.tight_layout()
plt.show()
```

### Feature Distribution


The goal variable ("Outcome") is used to create a series of subplots with KDE plots for each feature in a diabetic dataset. The target variable is not included in the list of column names for all the features in the variable features. The matplotlib.pyplot library's subplots function is used to produce a figure with 4 rows and 2 columns. Every feature is iterated through in a loop, which produces a KDE plot for each feature in the appropriate subplot. The KDE plot is made with different colors for the two target classes using the sns.kdeplot function from the seaborn library. Each subplot's title is set using the set_title function, and the subplot spacing is changed using plt.subplots_adjust. 

```python
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(13, 18),dpi=200)
for i, feature in enumerate(features):
    row = i // 2
    col = i % 2
    sns.kdeplot(data = diabetes_df, x = diabetes_df[feature], hue = diabetes_df["Outcome"], linewidth = 2, ax = axes[row, col],fill = True, palette=['#72b7b2', '#A52A2A'])
    axes[row, col].set_title(f'Feature Distribution by Target for {feature}')
    plt.subplots_adjust(hspace=0.3)
plt.show()
```
